```python
category_mapping = {
    "Fiction": "Fiction",
    "Juvenile Fiction": "Children's fiction",
    "Biography & Autobiography" : "Nonfiction",
    "History": "Nonfiction",
    "Literary Criticism": "Nonfiction",
    "Religion": "Nonfiction",
    "Philosophy":"Nonfiction",
    "Comics & Graphic Novels":"Fiction",
    "Drama":"Fiction",
    "Juvenile Nonfiction":"Children's Nonfiction",
    "Science":"Nonfiction",
    "Poetry":"Fiction",

}

books["simple_category"]=books["categories"].map(category_mapping)
```

```python
books
```

```python
from transformers import pipeline
fiction_categories = ["Fiction", "Nonfiction"]
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
sequence = books.loc[books['simple_category']=="Fiction","description"].reset_index(drop=True)[0]
```

```python
classifier(sequence,fiction_categories)
# The scores respresent the probability that the description belongs to that particular category
# in this case, the description belongs to "Fiction" with 84% probability
```

```python
import numpy as np
max_index = np.argmax(classifier(sequence,fiction_categories)["scores"])
max_label = classifier(sequence,fiction_categories)["labels"][max_index]
print(max_index)
print(max_label)
```

```python
def generate_predictions(sequence, categories):
    predictions = classifier(sequence, categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm
```

```python
actual_cats = []

predicted_cats = []

for i in tqdm(range(0,300)):
    sequence = books.loc[books['simple_category']=="Fiction","description"].reset_index(drop=True)[i]
    actual_cats.append("Fiction")
    predicted_cats.append(generate_predictions(sequence,fiction_categories))
for i in tqdm(range(0,300)):
    sequence = books.loc[books['simple_category']=="Nonfiction","description"].reset_index(drop=True)[i]
    actual_cats.append("Nonfiction")
    predicted_cats.append(generate_predictions(sequence,fiction_categories))

    
```

```python
x = 0
for i in range(len(actual_cats)):
    if actual_cats[i]==predicted_cats[i]:
        x+=1
print(x/len(actual_cats))
# We got a 0.78 accuracy in zero-shot classification
```
